```python
import getpass, os

USERNAME = "LeilaAyman"        # your GitHub username
OWNER = "LeilaAyman"           # repo owner (same as username)
REPO = "visualization-repo"    # ONLY the repo name

# Ask for your token safely
token = getpass.getpass("Enter your GitHub token: ")

# Build authenticated URL
repo_url = f"https://{USERNAME}:{token}@github.com/{OWNER}/{REPO}.git"

# Export for shell
os.environ["REPO_URL"] = repo_url
os.environ["REPO_NAME"] = REPO

print("Repo URL is ready.")
```

```python
# Remove old clone (if any)
!rm -rf $REPO_NAME

# Clone using token
!git clone "$REPO_URL"

# IMPORTANT: cd without quotes, without $ inside a string
%cd visualization-repo
```
